# Basic Linear Regression Model on India 

```python
india_case.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#coverting string date to date_time
import datetime as dt
india_case['date']=pd.to_datetime(india_case['date'])
india_case.head(10)
```

```python
#converting date_time to ordinal
india_case['date']=india_case['date'].map(dt.datetime.toordinal)
india_case.head()
```

```python
#getting dependent and independent variables
x=india_case['date']
y=india_case['total_cases']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
import numpy as np
lr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
india_case.tail()
```

```python
y_pred=lr.predict(np.array(x_test).reshape(-1,1))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(x_test,y_pred)
```
